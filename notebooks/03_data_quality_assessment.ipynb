## Assess Missing Values

```python
events_df.isnull().sum()
```

```python
missing = (
    events_df
    .isnull()
    .sum()
    .sort_values(ascending=False)
)

missing
```

```python
missing_percentage = (
    events_df
    .isnull()
    .mean()
    .mul(100)
    .round(2)
    .sort_values(ascending=False)
)

missing_percentage
```

```python
high_missing = missing_percentage[missing_percentage > 90]

high_missing
```
